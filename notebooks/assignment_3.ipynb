### Question 1

```python
def categorize_weather(temp):
    if temp < 0 :
        print(f"The temperature is Freezing")
    elif 0 <= temp <= 15:
        print(f"The temperature is Cold")
    elif 16 <= temp <= 30 :
        print(f"The temperature is Warrm")
    else:
        print(f"The temperature is Hot")
        
```

```python
# function call
categorize_weather(-2)
categorize_weather(5)
categorize_weather(45)
categorize_weather(19)
```

### Question 2

```python
def calculate_total(x: list):
    ''' 
    Calculate and total a list of item prices
    arguments :
         x(list) : list of item prices [0,100]
    returns :
         total : final price to pay         
    '''    
    total = 0

    # Loop to calculate total
    for price in x:
        total =  total + price

    # Apply discount AFTER summing all prices
    if total > 100:
        discount = total * 0.10
        discount = round(discount, 2)
        total = round(total - discount, 2)
        print(f"10% Discount applied: -${discount}")

    return total
```

```python
# Call function
price = [20,50,40,10]
calculate_total(price)
```

```python
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie', 'Diana'], 'Salary': [50000, 60000, 55000, 70000]}
df = pd.DataFrame(data)
df.head()
```

```python
# define function
def get_bonus(salary : int):
    if salary < 60000:
         return round(salary * 0.05,0)
    else :
       return round(salary * 0.02,0)
```

```python
# call function
df['Bonus_Amount'] = df['Salary'].apply(get_bonus)
print(df)
```
