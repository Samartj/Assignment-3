### Question 1

```python
def categorize_weather(temp):
    if temp < 0 :
        print(f"The temperature is Freezing")
    elif 0 <= temp <= 15:
        print(f"The temperature is Cold")
    elif 16 <= temp <= 30 :
        print(f"The temperature is Warrm")
    else:
        print(f"The temperature is Hot")
        
```

```python
# function call
categorize_weather(-2)
categorize_weather(5)
categorize_weather(45)
categorize_weather(19)
```

### Question 2

```python
def calculate_total(x: list):
    ''' 
    Calculate and total a list of item prices
    arguments :
         x(list) : list of item prices [0,100]
    returns :
         total : final price to pay         
    '''    
    total = 0

    # Loop to calculate total
    for price in x:
        total =  total + price

    # Apply discount AFTER summing all prices
    if total > 100:
        discount = total * 0.10
        discount = round(discount, 2)
        total = round(total - discount, 2)
        print(f"10% Discount applied: -${discount}")

    return total
```

```python
# Call function
price = [20,50,40,10]
calculate_total(price)
```

```python
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie', 'Diana'], 'Salary': [50000, 60000, 55000, 70000]}
df = pd.DataFrame(data)
df.head()
```

```python
# define function
def get_bonus(salary : int):
    if salary < 60000:
         return round(salary * 0.05,0)
    else :
       return round(salary * 0.02,0)
```

```python
# call function
df['Bonus_Amount'] = df['Salary'].apply(get_bonus)
print(df)
```

### Question 6

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
books_df = pd.DataFrame({'Student': ['Sam', 'Mia', 'Leo', 'Zoe'], 'Books_Read': [12, 25, 9, 18]})
plt.figure(figsize = (12,4))
sns.barplot(data = books_df, x = 'Student', y = 'Books_Read' )
plt.title("Reading Competition Results")
plt.xlabel("Students")
plt.ylabel("Books Read")
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```

###  Question 7

```python
import numpy as np
np.random.seed(1)
df = pd.DataFrame({'Height' : np.random.normal(loc = 170,scale =10,size =500)})
plt.figure(figsize = (8,4))
ax = sns.histplot(data = df , x = 'Height', bins = 10)
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2,p.get_height(),f'{int(p.get_height())}', ha = 'center', va ='bottom')
plt.title("Height Distribution")
plt.xlabel("")
plt.ylabel("")
plt.yticks([])
plt.gca().spines[['top','right']].set_visible(False)
plt.show()
```

```python
weather_df = pd.DataFrame({'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'], 'Temp': [22, 24, 19, 20, 25, 28, 30]})
plt.figure(figsize = (15,5))
sns.lineplot(data = weather_df , x = 'Day' , y = 'Temp', label = 'Temp' , linestyle = '-',marker ='o', markersize = 5)
plt.title(" Temperature Over One Week")
plt.xlabel("")
plt.ylabel("")
plt.gca().spines[['top','right']].set_visible(False)
plt.legend(bbox_to_anchor = (1,1), loc = 'upper left' )
plt.show()
```

```python
feedback_df = pd.DataFrame({'Rating': ['Good', 'Bad', 'Good', 'Excellent', 'Good', 'Bad', 'Excellent', 'Excellent', 'Good']})
plt.figure(figsize = (8,4))
ax = sns.countplot(data = feedback_df, x = 'Rating' ,order = feedback_df['Rating'].value_counts().index)
for p in ax.patches:
    height = p.get_height()  
    x = p.get_x() + p.get_width() / 2 
    y = height 
    ax.text(x, y, str(height), ha='center', va='bottom')
plt.title("Customer Feedback ")  
plt.xlabel('')
plt.ylabel('Ratings')
plt.yticks([])
plt.gca().spines[['top','left','right']].set_visible(False)
plt.show()
```
